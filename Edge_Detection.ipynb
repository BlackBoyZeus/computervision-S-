```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image from a URL or your local drive
# Replace 'image_url_or_file_path' with the URL or path to your image
image_url_or_file_path = '/content/ring.png'
image = cv2.imread(image_url_or_file_path, cv2.IMREAD_GRAYSCALE)

# Perform Canny edge detection
edges = cv2.Canny(image, threshold1=100, threshold2=200)

# Find contours in the edge map
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Create a blank mask to draw each contour separately
height, width = edges.shape[:2]
mask = np.zeros((height, width), dtype=np.uint8)

# Draw each contour on the mask
for contour in contours:
    cv2.drawContours(mask, [contour], 0, 255, thickness=cv2.FILLED)

# Display the original image and the individual masks
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(mask, cmap='gray')
plt.axis('off')
plt.title('Segmented Masks')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import random

# Load the image from a URL or your local drive
# Replace 'image_url_or_file_path' with the URL or path to your image
image_url_or_file_path = '/content/ring.png'
image = cv2.imread(image_url_or_file_path, cv2.IMREAD_GRAYSCALE)

# Display the original image
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.title('Original Image')
plt.show()

# Perform Canny edge detection
edges = cv2.Canny(image, threshold1=100, threshold2=200)

# Print the number of edges
num_edges = np.count_nonzero(edges)
print('Number of edges:', num_edges)

# Print the average edge strength
edge_strength = np.mean(edges)
print('Average edge strength:', edge_strength)

# Find contours in the edge map
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Create a blank mask to draw each contour separately
height, width = edges.shape[:2]
mask = np.zeros((height, width, 3), dtype=np.uint8)

# Generate a unique random color for each contour
def random_color():
    return (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))

# Create a dictionary to map colors to region labels
region_labels = {}
label_counter = 0

# Draw each contour with a unique random color on the mask and assign a label
for i, contour in enumerate(contours):
    color = random_color()
    label = chr(ord('A') + label_counter)  # Use alphabets as labels (A, B, C, ...)
    region_labels[label] = color
    label_counter += 1

    print(f"Drawing contour {i + 1} with label {label} and color: {color}")
    cv2.drawContours(mask, [contour], 0, color, thickness=cv2.FILLED)

# Display the colorful segmentation masks
plt.imshow(mask)
plt.axis('off')
plt.title('Colorful Segmentation Masks')
plt.show()

# Print the region labels and their corresponding colors
print("Region Labels and Colors:")
for label, color in region_labels.items():
    print(f"{label}: {color}")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import random

# Load the image from a URL or your local drive
# Replace 'image_url_or_file_path' with the URL or path to your image
image_url_or_file_path = '/content/ring.png'
image = cv2.imread(image_url_or_file_path, cv2.IMREAD_GRAYSCALE)

# Display the original image
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.title('Original Image')
plt.show()

# Perform Canny edge detection
edges = cv2.Canny(image, threshold1=100, threshold2=200)

# Print the number of edges
num_edges = np.count_nonzero(edges)
print('Number of edges:', num_edges)

# Print the average edge strength
edge_strength = np.mean(edges)
print('Average edge strength:', edge_strength)

# Find contours in the edge map
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Create a blank mask to draw each contour separately
height, width = edges.shape[:2]
mask = np.zeros((height, width, 3), dtype=np.uint8)

# Generate a unique random color for each contour
def random_color():
    return (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))

# Create a dictionary to map colors to region labels
region_labels_colors = {}
for i, contour in enumerate(contours):
    label = chr(ord('A') + i)  # Use alphabets as labels (A, B, C, ...)
    color = random_color()
    region_labels_colors[label] = color

    print(f"Drawing contour {i + 1} with label {label} and color: {color}")
    cv2.drawContours(mask, [contour], 0, color, thickness=cv2.FILLED)

# Function to get the color of a specific region based on its label
def get_color_for_label(label):
    return region_labels_colors.get(label, None)

# Display the colorful segmentation masks
plt.imshow(mask)
plt.axis('off')
plt.title('Colorful Segmentation Masks')
plt.show()

# Example usage: Retrieving colors of specific labels
label_to_check = 'D'
color_of_label = get_color_for_label(label_to_check)
if color_of_label:
    print(f"The color of region {label_to_check} is {color_of_label}")
else:
    print(f"Region {label_to_check} not found.")

# Example usage for multiple labels:
labels_to_check = ['A', 'B', 'C']
for label in labels_to_check:
    color_of_label = get_color_for_label(label)
    if color_of_label:
        print(f"The color of region {label} is {color_of_label}")
    else:
        print(f"Region {label} not found.")
```
